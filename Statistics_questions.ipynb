### Q1. State the CLT


 

The CLT states that if the mean and standard deviation of population parameter is finite then , sampling distribution of sample means will have a normal distribution with the mean equal to the population mean and the standard deviation equal to the standard deviation of the population  divided by the size of each samples irrespective of the type of the population distribution(binomial, skewed, etc)


Sample size >=30 would be preferrable to get normal distribution of sample means.


- CLT is used for estimating the population parameter(either using point estimation or interval estimataion) using sampling distribution of sample statistics.

- For eg : Need to find the avg. height of Indians. here CLT comes into play , which uses approximation theory to estimate population parameter.

###  b) PDF :

The probability density function  defines the probability distribution of a continuous random variable. When the probability density function is graphically plotted, the area under the curve will indicate the interval in which the variable will fall.

For eg :
Find the prob of 3 cm of rain:
HERE, we can define  the probability of exactly 3 cm rain, as on  particular day it may not be exactly 3, it can be 3.3, 3.5, 3.001, 2.9999 — so the probability of 3 cm would lie in certain range   (2.9 <3< 3.1) and that area that we find within that range is called PDF.


### c) PMF:
The probability distribution of a Discrete Random Variable is called as Probability Mass Function or PMF. 

- All probabilities are positive : P(x) ≥ 0. 
- The sum of all probabilities is 1. So ∑P(x) = 1.


```python
### PMF eg :
## PMF example : discrete distribution of data like that in a roll of unbiased dice.
## The sample : [1,2,3,4,5,6].
##The probability of each of the numbers appearing will be 1/6 or 0.167.
##This is a uniform distribution since the probabilities of all the values in the sample are the same.
## So the PMF will look something like this:

import seaborn as sns
import  pandas as pd
sample1 = [1,2,3,4,5,6] ## Discrete random variable for rolling a dice
df2 = pd.DataFrame(sample1,columns=['DiceRolls'])
probabilities = df2['DiceRolls'].value_counts(normalize=True)  ## normal= true : gives persentage counts  
print(probabilities)
sns.barplot(sample1, probabilities, color='blue')
```

solution :



Here , sample size is 80 and only one sample is given and we are assuming it is normal distribution so we do Z test

- Ho= There is no difference between popultion mean and sample mean 
- Ha = There is a significant  difference between sample mean and population mean

```python
#find z calculated 
 
import numpy as np
zcal=(370-400)/(30/np.sqrt(30))
print(zcal)
if (zcal>1.96 or zcal<(-1.96)):  # here 1.96 is z crtical at @ 0.05% (1-0.95)
    print("Ho will be rejected ")
else:
    print("Ho will be rejected ")

```

## Q-8



Solution:

Since we have categorical data so we use chisquare test:


Let 


- Ho: There is no difference between observed values and expected values

- Ha : There is a significant difference between observed values and expected values.

```python
import pandas as pd
import numpy as np
df1=pd.DataFrame([[60,50,30,20],[30,20,15,10]],columns=["A","B","C","D"],index=["owned 1 dog","owned 2 dog"])
df1 # observed Table
```

```python
df1.values
```

```python
import scipy.stats 
chi_vals=scipy.stats.chi2_contingency(df1) ### in backend it calculates all the computation like expected values,pvalues etc
chi_vals  # it returns ch square cofficient,p value, degree of freedom respectively
```

```python
if chi_vals[1]<0.05:
    print("Ho will be rejected")
else:
    print("Ho will be accepted  ")
print("degree of freedom is :",chi_vals[2])
```

means there is no significant difference between obsereved values and expected values  and degree of freedom is 3
